Visualize cdf to help understand the distribution of p-value

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import scipy.stats as stats
```

```python
# for linear samples
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)
```

```python
# Plot the standard normal distribution
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Standard Normal Distribution', color='blue')
```

```python
# cdf of the standard normal distribution
z = norm.cdf(x)
plt.plot(x, z, label='CDF', color='red')
```

```python
# two tails
z_tt = 2 * (1 - stats.norm.cdf(abs(x)))
plt.plot(x, z_tt, label='CDF', color='red')
```

```python
# for samples follows normal distribution
data = np.random.normal(0, 1, 1000)

p = norm.cdf(data)

plt.hist(p, bins=10, alpha=0.6, color='g')
```

## 2D samples

```python
# two dimensions
z = np.random.multivariate_normal([0, 0], [[1, 0], [0, 1]], 1000)
p = norm.cdf(z)

# plt.scatter(z[:, 0], z[:, 1], label='Sample', color='blue')

fig, ax = plt.subplots(1, 2)

ax[0].scatter(z[:, 0], z[:, 1], label='Sample', color='blue')
ax[0].set_title('2D Normal Distribution')

ax[1].scatter(p[:, 0], p[:, 1], label='Sample', color='blue')
ax[1].set_title('cdf of 2D Normal Distribution')
```

```python
# Calculate p-values for each sample
p_values_x = 2*(1 - norm.cdf(np.abs(z[:, 0])))
p_values_y = 2*(1 - norm.cdf(np.abs(z[:, 1])))

plt.scatter(p_values_x, p_values_y, c='blue', label='Samples')
plt.scatter(p[:, 0], p[:, 1], c='red', label='Samples')
```
